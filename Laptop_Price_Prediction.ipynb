# Building Model

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_absolute_percentage_error
```

```python
lnr = LinearRegression()
lsr = Lasso()
rdr = Ridge()
svr = SVR()
knr = KNeighborsRegressor()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
gbr = GradientBoostingRegressor()
```

```python
models = [lnr,lsr,rdr,svr,knr,dtr,rfr,gbr]
```

```python
x_train.dtypes
```

```python
for i in models:
  s1 = make_column_transformer((StandardScaler(),[2,4,7,9,10,11]),
                               (OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[0,1,3,5,6,8]),
                                remainder='passthrough')
  s2 = i
  pipe = Pipeline([('s1',s1),('s2',s2)])
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  print(i)
  print('R2 Score:',r2_score(y_test,y_pred))
  print('MAE:',mean_absolute_error(y_test,y_pred))
  print('MAPE:',mean_absolute_percentage_error(y_test,y_pred),'\n')
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Hyperparameter Tuning on Ridge Regressor

params = {'alpha':[1,5,10,20,30,35,40,45,50,55,100]}
gs = GridSearchCV(rdr,params,cv=5)
s2 = gs
pipe = Pipeline([('s1',s1),('s2',s2)])
pipe.fit(x_train,y_train)
print(gs.best_params_)
```

```python
# Hyperparameter Tuning on Random Forest Regressor

params = {'n_estimators':[75,100,125,150],'max_depth':[3,5,7,None],
          'max_features':['sqrt','log2',None],'max_leaf_nodes':[3,5,7,None],
          'min_samples_leaf':[1,2,4],'min_samples_split':[2,4,6]}
gs = GridSearchCV(rfr,params,cv=5)
s2 = gs
pipe = Pipeline([('s1',s1),('s2',s2)])
pipe.fit(x_train,y_train)
print(gs.best_params_)
```

```python
# Fitting Random Forest Regressor with best parameters

s2 = RandomForestRegressor(n_estimators=150,max_features='log2',
                           max_depth=None,max_leaf_nodes=None,
                           min_samples_leaf=1,min_samples_split=2)
pipe = Pipeline([('s1',s1),('s2',s2)])
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score:',r2_score(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MAPE:',mean_absolute_percentage_error(y_test,y_pred),'\n')
```

```python
# Hyperparameter Tuning on Gradient Boosting Regressor

params = {'n_estimators':[75,100,125,150],'max_depth':[3,5,7,None],
          'max_features':['sqrt','log2',None],'max_leaf_nodes':[3,5,7,None],
          'learning_rate':[.001,.01,.1,10],'subsample':[.25,.5,1]}
gs = GridSearchCV(gbr,params,cv=5,n_jobs=-1)
s2 = gs
pipe = Pipeline([('s1',s1),('s2',s2)])
pipe.fit(x_train,y_train)
print(gs.best_params_)
```

```python
# Fitting Gradient Boosting Regressor with best parameters

s2 = GradientBoostingRegressor(n_estimators=150,max_features='sqrt',
                               max_depth=7,max_leaf_nodes=None,
                               learning_rate=.1,subsample=1)
pipe = Pipeline([('s1',s1),('s2',s2)])
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score:',r2_score(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MAPE:',mean_absolute_percentage_error(y_test,y_pred),'\n')
```
